```python
import numpy as np
import pandas as pd
import itertools
import matplotlib.pyplot as plt

from glob import glob as glob

from scipy.spatial import distance

pd.options.display.max_rows = 200
```

```python
def load_sub_order_df(sub_num):
  fpathL = glob('data/behav/silvy_buckets/sub%iday2/%i_viewing_*trials.csv'%(100+sub_num,sub_num))
  fpathL.sort()
  orderdf = pd.read_csv(fpathL[-1]).iloc[:12]
  orderdf = orderdf.loc[:,('northOrSouth','stimFile1a')]
  orderdf.columns = ['NorS','wed_id']
  orderdf.NorS = orderdf.NorS.str.split('.').str[0].str[0].str.capitalize()
  orderdf.wed_id = orderdf.wed_id.str.split('-').str[-1].str.split('.').str[0]
  orderdf.loc[:,'wed_num'] = np.arange(12)
  orderdf.loc[:,'sub_num'] = sub_num
  return orderdf
```

```python
sub_num=33
glob('data/behav/silvy_buckets/sub%iday2/%i_viewing_*trials.csv'%(100+sub_num,sub_num))
```

```python
load_sub_order_df(32)
```

```python
sub_orderdfL = []
for sub_num in np.arange(45):
  try:
    sub_orderdfL.append(load_sub_order_df(sub_num))
  except:
    print(sub_num)
    continue
```

```python
orderdf = pd.concat(sub_orderdfL,ignore_index=True)
```
