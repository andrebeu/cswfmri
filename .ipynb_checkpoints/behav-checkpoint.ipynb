# recall transitions

```python
def get_recall_order(sub_wed_df):
  """
  given the df of a given subject on a given wedding
  returns np array with recall order (i.e. removing number of TRs)
  """
  sub_wed_behav = sub_wed_df['behavior'].values
  mask = sub_wed_behav[:-1] != sub_wed_behav[1:]
  recall_order = sub_wed_behav[:-1][mask]
  return recall_order

def recall_order_to_transition(recall_order):
  if len(recall_order)<1: return np.array([])
  L = []
  for i in range(len(recall_order)-1):
    L.append(recall_order[i:i+2])
  return np.array(L)
```

```python
recall_order
```

```python
transitionL = []
for (sub,wed),sub_wed_df in behav_df.groupby(['sub','wedding']):
  recall_order = get_recall_order(sub_wed_df)
  print(sub,wed,np.nonzero(recall_order))
  recall_order = np.nonzero(recall_order)[0]
  transition = recall_order_to_transition(recall_order)
  if len(transition)==0: continue
  transitionL.append(transition)
transition_arr = np.concatenate(transitionL)
```

```python
# subjects and weddings where schema was mentioned
schema_mention_D = {}
grouping_obj = behav_df[behav_df['behavior']==1].groupby('sub')
for sub_n_,df_ in grouping_obj:
  print('sub:',sub_n_, df_['wedding'].unique())
  schema_mention_D[sub_n_] = df_['wedding'].unique()
  
zero_schema_mention_subs = [item for item in ALL_SUB_NS if item not in list(schema_mention_D.keys())]
```

```python
plt.figure(figsize=(20,5))
ax = plt.gca()
n_weddings_schema_mention = [len(i) for i in schema_mention_D.values()]
df_schema_mention = pd.DataFrame(data=n_weddings_schema_mention,
                                 index=schema_mention_D.keys(),
                                 columns=['mentions'])

df_schema_mention.sort_values('mentions',ascending=False).plot.bar(ax=ax)
ax.set_title('#weddings where schema is mentioend')
ax.set_xlabel('subject')
# plt.savefig('figures/schema_count')
```
